```python
import pandas as pd
import numpy as np
from statsmodels.tools.eval_measures import rmse
from sklearn.metrics import mean_absolute_percentage_error
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
```

```python
# Run the stored data from the Data_Preparation.ipynb
%run ..//Data_Preparation.ipynb
```

```python
#Using the period 1990-2015 for training and 2016-2020 for testing
gdp_train = gdp[(gdp['Time'] >= '1990-01-01') & (gdp['Time'] < '2016-01-01')]

gdp_test = gdp[(gdp['Time'] >= '2016-01-01') & (gdp['Time'] < '2020-01-01')]

#Since gdp is not stationary, we will take the first difference
gdp_train['Value_diff'] = gdp_train['Value'].diff().fillna(0)

#Printing the length of the training and testing data
print(len(gdp_train))
print(len(gdp_test))
```

```python
#Extract the target variable
endog_train = gdp_train['Value_diff'] #The model is trained on the first difference
endog_test = gdp_test['Value'] #The model is tested on the original values

# Define hyperparameters grid for grid search
param_grid = {
    'max_depth': [4, 5, 6], # The maximum depth of the tree
    'min_child_weight': [5, 10, 15], # The minimum sum of weights of all observations required in a child
    'gamma': [0.5, 1, 1.5], # The minimum loss reduction required to make a further partition on a leaf node of the tree
    'subsample': [0.8, 1.0], # The fraction of observations to be randomly sampled for each tree
    'colsample_bytree': [0.8, 1.0] # The fraction of columns to be randomly sampled for each tree
}

#Setting the seed for reproducibility
seed = 42

# Instantiating the XGBoost Regressor model
xgb_reg = xgb.XGBRegressor(objective='reg:squarederror', seed=seed)

# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=xgb_reg, # The model to be optimized
                           param_grid=param_grid, # The grid of hyperparameters
                           cv=3, # The cross-validation strategy (3-fold)
                           scoring='neg_mean_squared_error', # The scoring metric (neg_mean_squared_error for regression)
                           verbose=1 # Verbosity level (1 for low verbosity, 2 for high verbosity)
                           )

# Fitting the model to the training data through GridSearch
grid_search.fit(np.arange(len(endog_train)).reshape(-1, 1), endog_train)

# Getting the best parameters
best_params = grid_search.best_params_
print("Best parameters found:", best_params)

# Get the best estimator
best_xgb_reg = grid_search.best_estimator_

# Forecasting the next 16 periods into the future
forecast_indices = np.arange(len(endog_train), len(endog_train) + 16).reshape(-1, 1)
y_pred = best_xgb_reg.predict(forecast_indices)

# Reversing the first difference
y_pred_norm = np.cumsum(y_pred) + gdp_train['Value'].iloc[-1]

# Calculating the RMSE
rmse_value = rmse(endog_test, y_pred_norm)
#Calculating the MAPE
mape_value = mean_absolute_percentage_error(endog_test, y_pred_norm)*100 #Multiplying by 100 to get the percentage

print("The RMSE of the model is:", rmse_value)
print("The MAPE of the model is:", mape_value)
```

```python
#Plotting the results against the actual values
plt.figure(figsize=(20,10))
plt.plot(gdp_train['Time'].to_numpy(), gdp_train['Value'].to_numpy(), label='Train GDP')
plt.plot(gdp_test['Time'].to_numpy(), y_pred_norm, label='Forecasted GDP', color='red')
plt.plot(gdp_test['Time'].to_numpy(), gdp_test['Value'].to_numpy(), label='Actual GDP')
plt.legend(loc='upper left')
plt.show()
```
